```python
import pandas as pd
```

```python
path = "Resources/purchase_data.csv"
file = pd.read_csv(path)
file
```

```python
#Player Count
#Display the total number of players

Total_players = len(file["SN"].value_counts())
total_table = pd.DataFrame({"Total Players":[Total_players]})
total_table
```

```python
#Purchasing Analysis (Total)
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results
#Display the summary data frame

unique_items = file["Item ID"].unique()
count_unique = len(unique_items)
avg_price = round(file["Price"].mean(),2)
avg_price = f"${avg_price:}"
num_purchases = len(file["Purchase ID"])
total_revenue = file["Price"].sum()
total_revenue = f"${total_revenue:,}"
summary1_df = pd.DataFrame({"Number of Unique items":[count_unique],
                            "Average price":[avg_price],
                            "Number of Purchases":[num_purchases],
                           "Total Revenue":[total_revenue]})
summary1_df
```

```python
#Gender Demographics
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

file_unique = file.drop_duplicates(subset="SN",keep="first",inplace=False)
file_gender = file_unique["Gender"].value_counts()
file_male = len(file_unique.loc[file_unique["Gender"] == "Male", :])
file_male_percent = round((file_male/Total_players)*100,2)
file_male_percent = f"{file_male_percent:}%"
file_female = len(file_unique.loc[file_unique["Gender"] == "Female", :])
file_female_percent = round((file_female/Total_players)*100,2)
file_female_percent = f"{file_female_percent:}%"
file_others = len(file_unique.loc[file_unique["Gender"] == "Other / Non-Disclosed", :])
file_others_percent = round((file_others/Total_players)*100,2)
file_others_percent = f"{file_others_percent:}%"
gender_table = pd.DataFrame({"Total Count":file_gender,
                            "Percentage of Players":[file_male_percent,file_female_percent,file_others_percent]})
gender_table
```

```python
#Purchasing Analysis (Gender)
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

file_gender_group = file.groupby(["Gender"])
file_gender_summary = file_gender_group["Purchase ID"].count()
file_gender_purchases = round(file_gender_group["Price"].mean(),2)
file_gender_purchases = '$' + file_gender_purchases.astype(str)
file_gender_totalpurchase = round(file_gender_group["Price"].sum(),2)
file_gender_totalpurchase = '$' + file_gender_totalpurchase.astype(str)
avg_male_purchase = file.loc[file["Gender"] == "Male", :]
avg_male_purchase = round((avg_male_purchase["Price"].sum())/file_male,2)
avg_male_purchase = f"${avg_male_purchase:}"
avg_female_purchase = file.loc[file["Gender"] == "Female", :]
avg_female_purchase = round((avg_female_purchase["Price"].sum())/file_female,2)
avg_female_purchase = f"${avg_female_purchase:}"
avg_others_purchase = file.loc[file["Gender"] == "Other / Non-Disclosed", :]
avg_others_purchase = round((avg_others_purchase["Price"].sum())/file_others,2)
avg_others_purchase = f"${avg_others_purchase:}"
file_summary_gender = pd.DataFrame({"Purchase Count":file_gender_summary,"Average Purchase Price":file_gender_purchases,
                                    "Total Purchase Value":file_gender_totalpurchase,
                                    "Avg Total Purchase per Person":[avg_female_purchase,avg_male_purchase,avg_others_purchase]})
file_summary_gender
```

```python
#Age Demographics
#Establish bins for ages
#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table

age_max = file["Age"].max()
age_min = file["Age"].min()
bins = [0,9,14,19,24,29,34,39,50]
group_names = ("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
file_unique["Age Range"] = pd.cut(file_unique["Age"],bins,labels=group_names)
Age_count = file_unique.groupby("Age Range")["Purchase ID"].count()
Age_percent = round((Age_count / Total_players)*100,2)
Age_percent = Age_percent.astype(str)+'%'
Age_summarytable = pd.DataFrame({"Total Count":Age_count,"Percentage of Players":Age_percent})
Age_summarytable
```

```python
#Purchasing Analysis (Age)
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

file["Age Range"] = pd.cut(file["Age"],bins,labels=group_names)
range_purchase_count = file.groupby("Age Range")["Price"].count()
range_purchase_avg = round(file.groupby("Age Range")["Price"].mean(),2)
range_purchase_avg = '$' + range_purchase_avg.astype(str)
range_purchase_value = round(file.groupby("Age Range")["Price"].sum(),2)
range_purchase_value = '$'+ range_purchase_value.astype(str)
range_avg_person = round(file.groupby("Age Range")["Price"].sum()/(Age_count),2)
range_avg_person = '$' + range_avg_person.astype(str)
range_table_summary = pd.DataFrame({"Purchase Count":range_purchase_count,"Average Purchase Price":range_purchase_avg,
                                    "Total Purchase Value":range_purchase_value,
                                    "Avg Total Purchase per Person":range_avg_person})
range_table_summary
```

```python
#Top Spenders
#Run basic calculations to obtain the results in the table below
#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

name_purchases = file.groupby("SN")["Purchase ID"].count()
name_avg_price = file.groupby("SN")["Price"].mean()
name_total_value = file.groupby("SN")["Price"].sum()
top_spenders = pd.DataFrame({"Purchase Count":name_purchases,"Average Purchase Price":name_avg_price,
                            "Total Purchase Value":name_total_value})
top_spenders_summary = top_spenders.sort_values(["Total Purchase Value"],ascending = False)
top_spenders_summary["Total Purchase Value"] = top_spenders_summary["Total Purchase Value"].map("${:.2f}".format)
top_spenders_summary["Average Purchase Price"] = top_spenders_summary["Average Purchase Price"].map("${:.2f}".format)
top_spenders_summary.head()
```

```python
#Most Popular Items
#Retrieve the Item ID, Item Name, and Item Price columns
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

item_purchase = file.groupby(["Item ID","Item Name"])["Purchase ID"].count()
item_total_value = file.groupby(["Item ID","Item Name"])["Price"].sum()
item_price = file.groupby(["Item ID","Item Name"])["Price"].mean()
item_purchase_summary =  pd.DataFrame({"Purchase Count":item_purchase,"Item Price":item_price,
                                       "Total Purchase Value":item_total_value})
item_purchase_table = item_purchase_summary.sort_values(["Purchase Count"],ascending = False)
item_purchase_table["Item Price"] = item_purchase_table["Item Price"].map("${:.2f}".format)
item_purchase_table["Total Purchase Value"] = item_purchase_table["Total Purchase Value"].map("${:.2f}".format)
item_purchase_table.head()
```

```python
#Most Profitable Items
#Sort the above table by total purchase value in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the data frame

item_purchase_table2 = item_purchase_summary.sort_values(["Total Purchase Value"],ascending = False)
item_purchase_table2["Item Price"] = item_purchase_table2["Item Price"].map("${:.2f}".format)
item_purchase_table2["Total Purchase Value"] = item_purchase_table2["Total Purchase Value"].map("${:.2f}".format)
item_purchase_table2.head()
```

The observable trends based on the above Data are as follows

2. Although the game has more male players, when it comes to purchasing items Female players tend to spend more as
